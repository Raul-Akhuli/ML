```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the sigmoid function

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
x = np.linspace(-3, 3, 200)
```

```python
# define points and there classes

x_points = np.array([-2, -1, 1, 2])
y_labels = np.array([0, 0, 1, 1])   # 0 for negative class and 1 for the positive class
```

```python
# define values for theta and theta0

parameters = [
    (1, 0),         # theta = 1 and theta0 = 0
    (-1, 0),        # theta = -1 and theta0 = 0
    (0.5, 0),       # theta = 0.5 and theta0 = 0
    (5, 0)          # theta = 5 and theta0 = 0,  likely the correct classifier
]
```

```python
# create a figure with 4 subplots

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs = axs.ravel()  # flatten the 2D array of axes to 1D for easy iteration

for i, (theta1, theta0) in enumerate(parameters):
    
    z = theta1 * x + theta0
    y = sigmoid(z)
    
    ax = axs[i]
    ax.plot(x, y)
    
    # Plot class points at y=0
    ax.scatter(x_points, np.zeros_like(x_points), color='red', label='Class -1')
    ax.scatter(y_labels, np.zeros_like(y_labels), color='blue', label='Class +1')
    
    # Decision boundary (0.5 probability)
    ax.axhline(0.5, linestyle='--')
    
    ax.set_title(f'Theta={theta1}, Theta0={theta0}')
    ax.set_xlim(-3, 3)
    ax.set_ylim(0, 1)
    ax.legend()

plt.tight_layout()
plt.show()
```
