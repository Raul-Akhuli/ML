```python
# one hot encoding using onehotencoder.

import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Building a dummy employee dataset.

data = {
    "employee_id": [10, 20, 15, 25, 30],
    "gender": ['M', 'F', 'F', 'M', 'F'],
    "remarks": ['good', 'nice', 'good', 'great','nice']
}

print("data before going to datafreame: ",data)

# convert into a pandas dataframe.
df = pd.DataFrame(data)

df
```

```python
df.dtypes
```

```python
# we only need catagorical columns so 'object' dtype columns.
catagorical_cols = df.select_dtypes(include=['object']).columns.tolist()
```

```python
# Initialize OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)  # need output in numpy array formate
# not sparse matrix formate.
```

```python
# now it's to apply one hot encoding on catagorical columns.

one_hot_encoded = encoder.fit_transform(df[catagorical_cols])
```

```python
# create a dataframe with one hot encoded column.
# we use get_feature_names_out to get the column names for one hot encoded columns.

one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(catagorical_cols))
one_hot_df
```

```python
# now I need to add both the dataframes 'df' and 'one_hot_df' together.

df_encoded = pd.concat([df, one_hot_df], axis=1) # I want column wise.
df_encoded
```

```python
# now drop the original catagorical columns.
df_encoded.drop(catagorical_cols, axis=1, inplace=True) # inplace=True means I want to change the original dataframe.
df_encoded
```
