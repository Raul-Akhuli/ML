```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# original XOR dataset in 2D
X = np.array([[1, 1], [-1, -1], [1, -1], [-1, 1]])
y = np.array([1, 1, -1, -1])

# plotting the 2D points
plt.figure(figsize=(8,4))
for i, label in enumerate(y):
    plt.plot(X[i, 0], X[i, 1], 'ro' if label == 1 else 'bo')

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Original XOR points in 2D')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# original XOR dataset in 2D
X = np.array([[1, 1], [-1, -1], [1, -1], [-1, 1]])
y = np.array([1, 1, -1, -1])
```

```python
# Apply the feature transformation to 3D: x3 = x1 * x2
X_transformed = np.hstack((X, (X[:, 0] * X[:, 1]).reshape(-1, 1)))

# create a ploty figure
fig = go.Figure()

# add the points to the plot
colors = ['red' if label == 1 else 'blue' for label in y]
fig.add_trace(go.Scatter3d(x=X_transformed[: , 0], y=X_transformed[: , 1], z=X_transformed[: , 2], mode='markers', marker=dict(color=colors, size=5)))

# define the separator line
xx, yy = np.meshgrid(np.linspace(-2, 2, 50), np.linspace(-2, 2, 50))
zz = 0 * xx
fig.add_trace(go.Surface(x=xx, y=yy, z=zz, opacity=0.5, colorscale='Greens'))

# set the layout for the better view
fig.update_layout(title='Transformed XOR points in 3D with separator', 
                  scene=dict(xaxis_title='X1', yaxis_title='X2', zaxis_title='X3 = X1 * X2'),
                  margin=dict(l=0, r=0, b=0, t=30))

fig.show()
```
