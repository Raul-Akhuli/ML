```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loss function

def f(x):
    return x**2 # x**2 means x to the power of 2, which is the same as x*x
```

```python
# Derivation of the loss function

def df(x):
    return 2*x
```

```python
# Gradient descent parameters

x_start = 9   # initial guess
learning_rate = 0.15
n_iterations = 10
```

```python
# For storing the iteration steps for visualization

x_steps = [x_start]
y_steps = [f(x_start)]
```

```python
# values for plotting the parabola

x = np.linspace(-10, 10, 400)
y = f(x)




for i in range(n_iterations):
    plt.figure(figsize=(8, 5))

    # plot the parabola
    plt.plot(x, y, label="f(x) = x^2")

    # calculate position on parabola
    current_x = x_steps[-1]
    current_y = y_steps[-1]

    # calculate the gradient
    gradient = df(current_x)

    # update the position (gradient descent step)
    next_x = current_x - learning_rate * gradient
    next_y = f(next_x)

    # save the new position
    x_steps.append(next_x)
    y_steps.append(next_y)

    # plot the current step path
    plt.scatter(x_steps, y_steps, color='green')
    plt.plot(x_steps, y_steps, linestyle='-', marker='o', color='blue', label="Gradient Descent Path")

    # plot the tangent line (derivative at current point)
    tangent_x = np.linspace(current_x - 1, current_x + 1, 100)
    tangent_y = current_y + gradient * (tangent_x - current_x)
    plt.plot(tangent_x, tangent_y, linestyle='--', color='red', linewidth=0.9,label="Tangent Line")



    plt.title(f'Gradient Descent on f(x) = x^2 (Iteration {i+1})')
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()
```
